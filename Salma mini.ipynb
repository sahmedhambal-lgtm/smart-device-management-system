```python
# Base class for all smart devices
class SmartDevice:
    def __init__(self, name, device_id, status="OFF"):
        self.name = name
        self._device_id = device_id
        self._status = status

    @property
    def device_id(self):
        return self._device_id

    @device_id.setter
    def device_id(self, value):
        if not value:
            print("Device ID cannot be empty.")
        else:
            self._device_id = value

    @property
    def status(self):
        return self._status

    @status.setter
    def status(self, value):
        if value not in ("ON", "OFF"):
            print("Invalid status value.")
        else:
            self._status = value

    def power_on(self):
        self.status = "ON"
        print(f"[{self.name}] switched ON")

    def power_off(self):
        self.status = "OFF"
        print(f"[{self.name}] switched OFF")

    def show_details(self):
        print(f"--- {self.name} ---")
        print(f"ID: {self._device_id} | Status: {self._status}")


class TempSensor(SmartDevice):
    def __init__(self, name, device_id, temp_celsius):
        super().__init__(name, device_id)
        self._temp = temp_celsius

    @property
    def temperature(self):
        return self._temp

    @temperature.setter
    def temperature(self, value):
        self._temp = value

    def report_temperature(self):
        print(f"{self.name} reading: {self._temp}°C")

    def show_details(self):
        super().show_details()
        print(f"Temperature: {self._temp}°C")


class SecurityCam(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self._recording = "OFF"

    @property
    def recording(self):
        return self._recording

    def begin_recording(self):
        self._recording = "ON"
        print(f"{self.name}: recording started")

    def end_recording(self):
        self._recording = "OFF"
        print(f"{self.name}: recording stopped")

    def show_details(self):
        super().show_details()
        print(f"Recording: {self._recording}")


class SmartLamp(SmartDevice):
    BRIGHTNESS_MIN = 0
    BRIGHTNESS_MAX = 100

    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self._brightness = 50

    @property
    def brightness(self):
        return self._brightness

    @brightness.setter
    def brightness(self, value):
        if self.BRIGHTNESS_MIN <= value <= self.BRIGHTNESS_MAX:
            self._brightness = value
        else:
            print("Brightness value out of allowed range.")

    def brighten(self, step):
        self._brightness = min(self._brightness + step, self.BRIGHTNESS_MAX)
        print(f"New brightness: {self._brightness}")

    def dim(self, step):
        self._brightness = max(self._brightness - step, self.BRIGHTNESS_MIN)
        print(f"New brightness: {self._brightness}")

    def show_details(self):
        super().show_details()
        print(f"Brightness: {self._brightness}")


# --- Create device instances ---
sensor = TempSensor("Kitchen Sensor", "TS21", 25)
camera = SecurityCam("Backyard Cam", "SC01")
lamp = SmartLamp("Living Room Lamp", "LS54")

devices = [sensor, camera, lamp]


def show_all_info():
    for d in devices:
        d.show_details()

def power_all_on():
    for d in devices:
        d.power_on()

def power_all_off():
    for d in devices:
        d.power_off()

def check_temperature():
    sensor.report_temperature()

def adjust_brightness():
    sub = input("Type 'up' to brighten or 'down' to dim: ").strip().lower()
    try:
        amt = int(input("Enter amount: "))
    except ValueError:
        print("Please enter a valid whole number.")
        return
    if sub == "up":
        lamp.brighten(amt)
    elif sub == "down":
        lamp.dim(amt)
    else:
        print("Unrecognized option.")

def toggle_recording():
    action = input("Type 'start' or 'stop': ").strip().lower()
    if action == "start":
        camera.begin_recording()
    elif action == "stop":
        camera.end_recording()
    else:
        print("Unrecognized option.")


menu_actions = {
    "1": show_all_info,
    "2": power_all_on,
    "3": power_all_off,
    "4": check_temperature,
    "5": adjust_brightness,
    "6": toggle_recording,
}

menu_text = """
1. Show device info
2. Power ON all devices
3. Power OFF all devices
4. Check temperature
5. Adjust lamp brightness
6. Toggle camera recording
7. Quit
"""

def run_menu():
    while True:
        print(menu_text)
        try:
            selection = input("Enter your choice: ").strip()
        except KeyboardInterrupt:
            print("\nGoodbye!")
            break

        if selection == "7":
            print("Goodbye!")
            break
        elif selection in menu_actions:
            menu_actions[selection]()
        else:
            print("That's not a valid option, try again.")

run_menu()
```
